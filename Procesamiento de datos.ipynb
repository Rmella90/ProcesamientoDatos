```python
#Now the challenge is to generate a code that generates the cleaning of the entire dataframe without the need for manual work
# for each column
#Ahora el desafio es generar un código que genere la limpieza de todo el dataframe sin necesidad el trabajo manuamente
#  por cada columna
```

```python
#We copy and adapt the methods used previously in an iteration function and filter the Order ID column


def generarDfLimpio(df): #Defino mi función
    dataframes = {}  # Genero un diccionario
    dataframe_names = []  # Lista para almacenar los nombres de los DataFrames
        
    i = 1

    for columna in df.columns:
        if columna != 'Order Id':  # Excluye la columna 'Order Id'
            df_temporal = df.dropna(subset=[columna])  # Eliminar filas con valores nulos en la columna actual
            df_temporal = df_temporal[['Order Id', columna]]  # Seleccionar solo las columnas 'Order Id' y la columna actual
            ultima_letra = columna[-1] #Tomo la última letra de cada columna (C, P o H)
            segment = {"C": "Consumer", "P": "Corporate", "H": "Home Office"}.get(ultima_letra, None) #Utilizo un diccionario para extraer el segmento de esa columna
            df_temporal["Segment"] = segment #Agrego ese segmento a toda la fila
            df_temporal["ship_mode"] = columna[:-2] #Genero otra columna que es ship mode, donde agrego el nombre de la columna extrayendo los dos últimos caracteres
        
            df_temporal = df_temporal.rename(columns={columna: 'Order Cost'})
            df_temporal = df_temporal.reset_index(drop=True)
            # Almacenar el DataFrame resultante en el diccionario
            
            df_name = f'DF_{i}'  # Nombre único para el DataFrame
            dataframes[df_name] = df_temporal
            dataframe_names.append(df_name)  # Agregar el nombre a la lista
            i += 1  # Incrementar el contador

    return dataframes, dataframe_names
```

```python
#use the function

dataframes, dataframe_names = generarDfLimpio(df)

# we vertically concatenate all df generated with the function
concatenated_df = pd.concat([dataframes[name][['Order Id', 'Order Cost', 'Segment']] for name in dataframe_names], axis=0, ignore_index=True)
```
